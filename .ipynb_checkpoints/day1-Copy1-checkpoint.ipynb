### Step 5: Data Preprocessing
- seperate the features and labels
- do the label encoding
- solve for data imbalance
- train test split
- feature scaling

```python
def preprocess(data, label):
    # seperate the features and labal
    X = df.drop('Outcome',axis = 1)
    y = df['Outcome']
    
    # solve data imablance
    sm = SMOTE()
    X, y = sm.fit_resample(X,y)
    
    # train test split
    X_train, X_test, y_train , y_test = train_test_split(X,y, random_state = 42, stratify = y, test_size = 0.20)
    
    return  X_train, X_test, y_train , y_test
    
    # startify will maintain the same number of data in train and test
```

```python
 X_train, X_test, y_train , y_test = preprocess(df, label = 'Outcome')
```

```python
# scale teh features

sc = StandardScaler()

x_train = sc.fit_transform(X_train)
x_test = sc.transform(X_test)
```

```python
# we have preperocessed the data
```

### Step 6: Fit and Evaluate ML Algorithms


```python
# create a metrics fucntion

def print_metrics(y_test, y_pred, model_name):
    print('Metrics for model...', model_name)
    print('')
    print('Accuracy Score = ', accuracy_score(y_test, y_pred))
    print('')
    print('Recall Score = ', recall_score(y_test, y_pred))
    print('')
    print('Presicion Score = ', precision_score(y_test, y_pred))
    print('')
    print('F1 Score = ', f1_score(y_test, y_pred))
    print('')
    print('ROC AUC Score = ', roc_auc_score(y_test, y_pred))
    print('Confusion Matrix')
    print(confusion_matrix(y_test, y_pred))
    print('Classification Report')
    print(classification_report(y_test, y_pred))
```

```python
# print and evaluate a KNN model

knn = KNeighborsClassifier()
knn.fit(x_train,y_train )
y_pred = knn.predict(x_test)
print_metrics(y_test, y_pred,'KNN')
```

```python
# Lets optimize the neighbours to improve by drowing model complexity curve

neighbors = np.arange(1,12)
train_accuracies = np.empty(len(neighbors))
test_accuracies = np.empty(len(neighbors))

# enumerate over the neighbours

for i,k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors= k)
    knn.fit(x_train, y_train)
    train_accuracies[i] = knn.score(x_train, y_train)
    test_accuracies[i] = knn.score(x_test, y_test)
    
    
# plot the model complexity curve
plt.title('Model Complexity Curves')
plt.plot(neighbors, train_accuracies, label = 'Training Metrics')
plt.plot(neighbors, test_accuracies, label = 'Testing Metrics')
plt.legend()
plt.show()
```

```python
plt.style.use('fivethirtyeight')
```

```python
# refit knn with

knn = KNeighborsClassifier(n_neighbors= 9)
knn.fit(x_train,y_train )
y_pred = knn.predict(x_test)
print_metrics(y_test, y_pred,'KNN')
```

```python
# fit all models togeather to get the best model to optimize

clfs = {
    'logreg':LogisticRegression(),
    'knn' : KNeighborsClassifier(),
    'Naive Bayes' : GaussianNB(),
    'Decision Tree' : DecisionTreeClassifier(),
    'rfc' : RandomForestClassifier(),
    'ABC' : AdaBoostClassifier(),
    'GBC' : GradientBoostingClassifier(),
    'SVM' : SVC(),
    'xgboost' : XGBClassifier()
}

models_report = pd.DataFrame(columns = ['Model Name', 'Accuracy','Recall','Precision','F1 Score'])

for clf, clf_name in list(zip(clfs.values(), clfs.keys())):
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    print('Fitting the model---', clf_name)
    t = pd.Series({
        'Model Name': clf_name, 'Accuracy': accuracy_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),'Precision': precision_score(y_test, y_pred),
        'F1 Score': f1_score(y_test, y_pred) })
    
    models_report = models_report.append(t, ignore_index = True)
model_report = models_report .sort_values(by = 'F1 Score', ascending = False)
model_report

```

### Step 7: Model Optimizations

- Random Forest came out the best and we will optimize it

```python
# fit a RFC to the data

rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_test)

print_metrics(y_test, y_pred, 'RFC')
```

```python
# check for overfitting

train_acc = rfc.score(x_train, y_train)
test_acc = rfc.score(x_test, y_test)
train_acc, test_acc
```

- There appears to be overfitting which has to be considered and could eb because we balanced the data or less number of data


```python
from sklearn.model_selection import RandomizedSearchCV

param_dict = {
    'n_estimators' :range(100, 1000, 50), 
    'min_samples_leaf' : range(1,5,1), 
    'min_samples_split' : range(2,10,2),
    'max_depth' : range(10,100,5),
    'max_features': ['auto','sqrt','log2'],
    'criterion': ['gini','entropy']    
}

n_folds = 10
rs = RandomizedSearchCV(estimator = rfc , param_distributions =param_dict,scoring = 'f1',
                       random_state = 42, n_jobs= -1, n_iter= 100, 
                       cv = n_folds)

rs.fit(x_train,y_train)
```

```python
# tuned rfc

rfc_tuned = RandomForestClassifier(criterion='entropy', max_depth=55, n_estimators=350)
rfc_tuned.fit(x_train, y_train)
y_pred = rfc.predict(x_test)

print_metrics(y_test, y_pred, 'Tuned RFC')
```

### Step 9 : Prepare for deployment for creating a pipeline

```python
from sklearn.pipeline import Pipeline
```

```python
sc = StandardScaler()
model = rfc_tuned

steps = [('scaler', sc),('model', model)]
pipeline = Pipeline(steps)

x_train, x_test, y_train, y_test = preprocess(df, label = 'Outcome')
# this is unscaled data

pipeline.fit(x_train, y_train)
y_pred = pipeline.predict(x_test)
print_metrics(y_test, y_pred, 'pipeline')
```
